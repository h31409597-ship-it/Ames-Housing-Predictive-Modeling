# Imports and Data Loading

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

df = pd.read_csv("featured_data.csv")
target = df['SalePrice'].values
```

Explanation: We import numpy for manual calculations and StandardScaler to verify our manual work. We convert the 'SalePrice' column into a NumPy array.

# Manual Mean and Standard Deviation

```python
manual_mean = np.sum(target) / len(target)
manual_std = np.sqrt(np.sum((target - manual_mean)**2) / len(target))

print(f"Manual Mean: {manual_mean}")
print(f"Manual Std: {manual_std}")
```

Explanation: Instead of using .mean(), we calculate the average by summing all values and dividing by the count. The standard deviation is calculated by finding the square root of the average squared deviations from the mean.

# Manual Standardization (Z-Score)

```python
column_data = df['Gr Liv Area'].values
z_manual = (column_data - np.mean(column_data)) / np.std(column_data)

scaler = StandardScaler()
z_sklearn = scaler.fit_transform(column_data.reshape(-1, 1)).flatten()

print(f"Difference: {np.sum(z_manual - z_sklearn)}")
```

Explanation: We standardize 'Gr Liv Area' manually using the formula $z = \frac{x - \mu}{\sigma}$. We then compare it with the StandardScaler output to prove they are identical (the difference should be near zero).

# Cosine Similarity between Extreme Records

```python
highest_idx = df['SalePrice'].idxmax()
lowest_idx = df['SalePrice'].idxmin()

numeric_cols = df.select_dtypes(include=[np.number]).columns
vec_high = df.loc[highest_idx, numeric_cols].values.reshape(1, -1)
vec_low = df.loc[lowest_idx, numeric_cols].values.reshape(1, -1)

similarity = cosine_similarity(vec_high, vec_low)
print(f"Cosine Similarity: {similarity[0][0]}")
```

Explanation: We find the most expensive and cheapest houses. We then calculate the Cosine Similarity between their feature vectors to see how similar they are in terms of direction/proportions, regardless of their price magnitude.

# Probability Estimation

```python
high_quality_mask = df['Overall Qual'] >= 8
threshold = df['SalePrice'].median()

high_quality_count = np.sum(high_quality_mask)
above_threshold_count = np.sum((df['Overall Qual'] >= 8) & (df['SalePrice'] > threshold))

probability = above_threshold_count / high_quality_count
print(f"Probability: {probability:.2f}")
```

Explanation: We estimate the probability: "If a house is high quality (8+), what is the chance its price is above the median?" This helps us understand the predictive power of the 'Quality' feature.
